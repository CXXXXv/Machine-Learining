```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = [0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
Y = [0, 0.6, 1, 1.5, 1.7, 2.5, 3, 3.5, 4, 4.8, 5]
```

```python
fig = plt.figure(figsize = (8, 8))
plt.xlim((-0.5, 5.5))
plt.ylim((-0.5, 5.5))
plt.scatter(X, Y, s=50, marker='x', color='red')
plt.show()
```

```python
theta0, theta1 = 2.5, 0
alpha = 0.05

for iteration in range(0, 300):
   
    J, dtheta0, dtheta1 = 0, 0, 0
    for i in range(len(X)):
        J = J + (theta0 + theta1 * X[i] - Y[i]) ** 2
        dtheta0 = dtheta0 + (theta0 + theta1 * X[i] - Y[i])
        dtheta1 = dtheta1 + (theta0 + theta1 * X[i] - Y[i]) * X[i]
    J /= (2 * len(X))
    dtheta0 /= len(X)
    dtheta1 /= len(X)
    
    if iteration % 20 == 0:
        x1 = 0; x2 = 5
        y1 = theta0 + theta1 * x1
        y2 = theta0 + theta1 * x2

        fig = plt.figure(figsize = (6, 6))
        plt.xlim((-0.5, 5.5))
        plt.ylim((-0.5, 5.5))
        plt.plot([x1, x2], [y1, y2], linewidth=2, color='blue')
        plt.scatter(X, Y, s=50, marker='x', color='red')
        plt.title('Iteratrion:' + '%d'% iteration + '  Cost:' + '%s' % round(J, 6))
        plt.show()
        
    theta0 = theta0 - alpha * dtheta0
    theta1 = theta1 - alpha * dtheta1
    
```
